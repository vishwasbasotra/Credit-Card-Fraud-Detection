# Exploratory Data Analysis

## Step 1. Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(2)
import tensorflow as tf
```

## Step 2. Importing Dataset

```python
dataset = pd.read_csv('dataset/creditcard.csv')
dataset.head(10)
```

## Step 3. Data Preprocessing

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
dataset['Amount'] = sc.fit_transform(dataset['Amount'].values.reshape(-1,1))
dataset.head(10)
```

```python
dataset.drop('Time', axis=1,inplace=True)
dataset.head(10)
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

## Step 4. Visualizing the Dataset

```python
fig = plt.figure(figsize=(20,70))
plt.suptitle('Histogram of Numerical Columns', fontsize=20)
for i in range(dataset.shape[1]):
    plt.subplot(10, 3, i+1)
    f = plt.gca()
    f.set_title(dataset.columns.values[i], fontsize = 16)
    
    vals = np.size(dataset.iloc[:,i].unique())
    if vals >= 100:
        vals = 100
    
    plt.hist(dataset.iloc[:,i], bins = vals, color = 'dodgerblue')
plt.tight_layout(rect=[0, 0.3, 1, 0.95])
```

```python
fraud = dataset['Class'].value_counts()[1]
nonfraud = dataset['Class'].value_counts()[0]
ratio = (fraud/nonfraud) * 100

print('Fraudulent: {fraud}, Non-Fraudulend: {nonfraud}'.format(fraud=fraud,nonfraud=nonfraud))
print('Ratio of Fraud to Non-Fraudulent: {fraud}/{nonfraud} = {ratio:.3f}%'.format(fraud=fraud,nonfraud=nonfraud,ratio=ratio))
```

```python
# fraud vs non-fraud transactions
plt.figure(figsize=(10,8))
sns.barplot(x=dataset['Class'].value_counts().index,y=dataset['Class'].value_counts(),color='blue')
plt.title('Fraudulent vs. Non-Fraudulent Transactions')
plt.ylabel('Count')
plt.xlabel('0:Non-Fraudulent, 1:Fraudulent')
plt.show()
```

```python
# correlation between features and the target
dataset.iloc[:,:-1].corrwith(dataset['Class']).plot(kind = 'bar',
                                          figsize=(20,10),
                                          color = 'lightcoral',
                                          fontsize = 15,
                                          rot = 45,
                                          grid = True
                                          )
plt.title('Correlation between Independent Variables and Dependent Variable', fontsize=20)
plt.show()
```

```python
# correlation matrix between the features
plt.figure(figsize=(25,15))
sns.set(font_scale=1.4)
sns.heatmap(dataset.corr(),  
            linewidths=.5, 
            cmap="RdBu_r",
            annot_kws={"size":10},
            cbar_kws={'shrink':.5}
           )
plt.xticks(rotation=45)
plt.title('Correlation Matrix', fontsize=25)
plt.show()
```

## Step 5. Splitting the Datset into Training set and Testing set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Step 6. Building the ANN(Artificial Neural Network)

### Initializing the ANN

```python
ann = tf.keras.models.Sequential()
```

### Adding the input layer and the first hidden layer

```python
ann.add(tf.keras.layers.Dense(units=16, activation='relu'))
```

### Adding 2nd hidden layer

ann.add(tf.keras.layers.Dense(units=24, activation='relu'))

### Adding a dropout layer

```python
ann.add(tf.keras.layers.Dropout(0.5))
```

### Adding 3rd hidden layer

```python
ann.add(tf.keras.layers.Dense(units=20, activation='relu'))
```

### Adding 4th hidden layer

```python
ann.add(tf.keras.layers.Dense(units=24, activation='relu'))
```

### Adding the output layer

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

## Step 7. Training the ANN

### Compiling the ANN
i.e., Applying Stochastic Gradient Descent

```python
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

### Training the ANN on the Training set

```python
ann.fit(X_train,y_train,batch_size=15,epochs=5)
```

```python
ann.summary()
```

## Step 8. Evaluation

```python
score = ann.evaluate(X_test,y_test)
```

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print('Accuracy: ',accuracy_score(y_test, y_pred))
print('\n\nClassification Report: \n\n',classification_report(y_test, y_pred))
plt.figure(figsize=(10,6))
sns.set(font_scale=1.4)
sns.heatmap(cm,  
            linewidths=.5,
            annot=True,
            fmt='g',
            cmap="RdBu_r"
           )
plt.title('Confusion Matrix', fontsize=14)
plt.show()
```

```python
y_pred = ann.predict(X)
y_pred = (y_pred > 0.5)
```

```python
cm = confusion_matrix(y, y_pred)
print('Accuracy: ',accuracy_score(y, y_pred))
print('\n\nClassification Report: \n\n',classification_report(y, y_pred))
plt.figure(figsize=(10,6))
sns.set(font_scale=1.4)
sns.heatmap(cm,  
            linewidths=.5,
            annot=True,
            fmt='g',
            cmap="RdBu_r"
           )
plt.title('Confusion Matrix', fontsize=14)
plt.show()
```
